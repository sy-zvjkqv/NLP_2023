```python
import pandas as pd
```

```python
df_test = pd.read_csv('/home/is/shuntaro-o/dev/persons_move_analysis/data/202271month_per_hour_geotaged_readGeocode_Tokyo.csv')
```

```python
df_test
```

```python
import spacy
nlp = spacy.load("ja_core_news_md")
```

```python
sentences = df_test.text.values
list_geo_Flag = []
for text in sentences:
    Flag = 0
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_=="GPE":
            Flag += 1
    #print(Flag)
    list_geo_Flag.append(Flag)
```

```python
list_geo_Flag
```

```python
list_geo_Flag=pd.Series(list_geo_Flag)
df_test=pd.concat([df_test,list_geo_Flag],axis=1)
df_test=df_test.rename(columns={0: 'Geo_Flag_multi'})
```
