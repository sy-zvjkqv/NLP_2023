```python
import folium

map = folium.Map(location=[33, 130], zoom_start=8)

osaka = [34.681667, 135.498611]
tokyo = [35.697778, 139.670833]

# デフォルトのピンを差す
folium.Marker(osaka, popup="長崎県庁", icon=folium.Icon(color='red')).add_to(map)
folium.Marker(tokyo, popup="長崎県庁", icon=folium.Icon(color='red')).add_to(map)



# 表示
map
```
